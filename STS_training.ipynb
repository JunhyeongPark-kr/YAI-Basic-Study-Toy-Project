```python
import math
import logging
from datetime import datetime

import torch
from torch.utils.data import DataLoader
from datasets import load_dataset
from sentence_transformers import SentenceTransformer, models, LoggingHandler, losses, util
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator
from sentence_transformers.readers import InputExample
```

```python
# logger
logging.basicConfig(
    format="%(asctime)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
    level=logging.INFO,
    handlers=[LoggingHandler()],
)
```

```python
pretrained_model_name = 'klue/roberta-base'
sts_num_epochs = 4
train_batch_size = 32

sts_model_save_path = 'output/training_sts-'+pretrained_model_name.replace("/", "-")+'-'+datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
```

# 1. Load Dataset & Preprocessing

## 1.1. KLUE-STS

```python
# load KLUE-STS Dataset
klue_sts_train = load_dataset("klue", "sts", split='train[:90%]')
klue_sts_valid = load_dataset("klue", "sts", split='train[-10%:]') # train의 10%를 validation set으로 사용
klue_sts_test = load_dataset("klue", "sts", split='validation')

print('Length of Train : ',len(klue_sts_train))
print('Length of Valid : ',len(klue_sts_valid))
print('Length of Test : ',len(klue_sts_test))
```

```python
klue_sts_train[0]
```

```python
def make_sts_input_example(dataset):
    '''
    Transform to InputExample
    '''
    input_examples = []
    for i, data in enumerate(dataset):
        sentence1 = data['sentence1']
        sentence2 = data['sentence2']
        score = (data['labels']['label']) / 5.0  # normalize 0 to 5
        input_examples.append(InputExample(texts=[sentence1, sentence2], label=score))

    return input_examples
```

```python
sts_train_examples = make_sts_input_example(klue_sts_train)
sts_valid_examples = make_sts_input_example(klue_sts_valid)
sts_test_examples = make_sts_input_example(klue_sts_test)
```

```python
# Train Dataloader
train_dataloader = DataLoader(
    sts_train_examples,
    shuffle=True,
    batch_size=train_batch_size,
)

# Evaluator by sts-validation
dev_evaluator = EmbeddingSimilarityEvaluator.from_input_examples(
    sts_valid_examples,
    name="sts-dev",
)

# Evaluator by sts-test
test_evaluator = EmbeddingSimilarityEvaluator.from_input_examples(
    sts_test_examples,
    name="sts-test",
)
```

# 2. Load Embedding Model

```python
# Load Embedding Model
embedding_model = models.Transformer(
    model_name_or_path=pretrained_model_name,
    max_seq_length=256,
    do_lower_case=True
)

# Only use Mean Pooling -> Pooling all token embedding vectors of sentence.
pooling_model = models.Pooling(
    embedding_model.get_word_embedding_dimension(),
    pooling_mode_mean_tokens=True,
    pooling_mode_cls_token=False,
    pooling_mode_max_tokens=False,
)

model = SentenceTransformer(modules=[embedding_model, pooling_model])
```

# 3. STS training

```python
# Use CosineSimilarityLoss
train_loss = losses.CosineSimilarityLoss(model=model)

# warmup steps
warmup_steps = math.ceil(len(sts_train_examples) * sts_num_epochs / train_batch_size * 0.1) #10% of train data for warm-up
logging.info("Warmup-steps: {}".format(warmup_steps))

# Training
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    evaluator=dev_evaluator,
    epochs=sts_num_epochs,
    evaluation_steps=int(len(train_dataloader)*0.1),
    warmup_steps=warmup_steps,
    output_path=sts_model_save_path
)
```

# 4. Evaluation

```python
# evaluation sts-test
test_evaluator(model, output_path=sts_model_save_path)
```

```python
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt')  # punkt tokenizer를 다운로드, 필요시 한 번만 실행

# 데이터셋에서 모든 텍스트 정보를 포함하는 데이터셋 생성 및 문장 단위 분할 함수
def create_sentence_dataset(papers_dataset):
    sentence_dataset = []
    for paper in papers_dataset:
        title = paper['title']['en']
        authors = " ".join([author['name'] for author in paper['authors']]) if 'authors' in paper and paper['authors'] is not None else ""
        journal = paper['journal']['en'] if 'journal' in paper and paper['journal'] is not None else ""
        year = str(paper['year'])
        abstract = paper['abstract']['en'] if 'abstract' in paper and paper['abstract'] is not None else ""
        keywords = ", ".join(paper['keywords']['en'].split(';')).strip() if 'keywords' in paper and paper['keywords'] is not None else ""

        body_texts = []
        for section in paper['body_text']:
            if 'text' in section:
                section_text = " ".join(section['text'])
                body_texts.append(section_text)
        body_text = " ".join(body_texts)

        full_text = f"{title}. {authors}. {journal}, {year}. {abstract} Keywords: {keywords}. {body_text}"

        # 문장 단위로 분할
        sentences = sent_tokenize(full_text)
        sentence_dataset.extend(sentences)

    return sentence_dataset


# 문장 데이터셋 생성
sentence_dataset = create_sentence_dataset(papers_dataset)

# 일부 결과 출력 (처음 5개 문장)
for sentence in sentence_dataset[:5]:
    print(sentence)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 데이터셋에서 모든 텍스트 정보를 포함하는 데이터셋 생성 함수
def create_dataset(papers_dataset):
    dataset = []
    for paper in papers_dataset:
        title = paper['title']['en']
        authors = paper['authors']
        journal = paper['journal']['en']
        year = str(paper['year'])
        abstract = paper['abstract']['en']
        keywords = ", ".join(paper['keywords']['en'].split(';')).strip()

        body_texts = []
        for section in paper['body_text']:
            # 'text' 키가 존재하는지 확인
            if 'text' in section:
                section_text = " ".join(section['text'])
                body_texts.append(section_text)
        body_text = " ".join(body_texts)

        full_text = f"{title} {authors} {journal} {year} {abstract} Keywords: {keywords} {body_text}"
        dataset.append(full_text)

    return dataset

# 가상의 사용자 입력 텍스트
# 실제 환경에서는 파일 입출력을 사용하여 텍스트 파일의 내용을 여기에 할당합니다.
input_text = "This study explores the impact of surrealism in modern fashion, focusing on the innovative use of symbols and decorations in clothing designs."

# 데이터셋 생성
dataset = create_dataset(papers_dataset)

# TF-IDF 벡터라이저를 사용하여 문서들의 TF-IDF 행렬 생성
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(dataset + [input_text])

# 코사인 유사도 계산
cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

# 가장 유사도가 높은 문서의 인덱스와 유사도 값 추출
most_similar_doc_index = cosine_similarities.argsort()[0][-1]
most_similar_doc_score = cosine_similarities[0, most_similar_doc_index]

print("가장 유사한 문서의 인덱스:", most_similar_doc_index)
print("유사도 점수:", most_similar_doc_score)
print(dataset)
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
import faiss

model = SentenceTransformer('/content/drive/MyDrive/YAI/STSmodel.h5')

# 데이터셋 구조 변경 및 문장 단위 벡터화
all_sentences = [sentence for paper in papers_dataset for sentence in paper]
sentence_embeddings = model.encode(all_sentences, convert_to_tensor=True)

# Faiss 인덱스 생성 및 벡터 추가
dimension = sentence_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(sentence_embeddings.cpu()))

###with open('input.txt', 'r', encoding='utf-8') as file:
    ###user_input = file.read()

user_sentence_vector = model.encode(input_text)  # 문장을 벡터로 인코딩
user_sentence_vector_2d = np.array([user_sentence_vector])  # 2차원 배열로 변환

# 유사한 문장 검색 실행
D, I = index.search(user_sentence_vector_2d, 10)  # 상위 10개 유사 문장 검색

# 결과 출력
print(D)
print(I)
```
